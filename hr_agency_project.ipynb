### Корректировка столбца `select_stages`
[Назад к содержанию](#section_id)

В столбце `select_stages` есть символы _x000D_ и \n - специальные символы, которые были вставлены в процессе экспорта данных из Excel в pandas DataFrame. Очистим данные от этих символов, чтобы читалось удобнее. 

```python
df.select_stages.unique()
```

```python
df['select_stages'] = df['select_stages'].astype(str).apply(openpyxl.utils.escape.unescape)
df['select_stages'] = df['select_stages'].str.replace('\n',',')
df['select_stages'] = df['select_stages'].str.replace('\r','')
```

```python
df.select_stages.value_counts().head(20)
```

На данном этапе уже можно сказать, что скрининг резюме и собеседование являются базовыми этапами. указанными в вакансиях.

```python
def clean_text(text):
    # Удаляем все символы, кроме букв и пробелов
    text = re.sub(r'[^а-яА-ЯёЁ\s]', '', text)
    # Преобразуем все буквы в нижний регистр
    text = text.lower()
    return text
```

```python
all_text = ' '.join(df['select_stages'].apply(clean_text))
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Отображаем облако слов
plt.figure(figsize=(10,  5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
words = all_text.split()

# Подсчитываем частоту каждого слова
word_counts = Counter(words)
word_counts
```

Резюме, собеседование, тестовое (задание), интервью - самые частые слова, встречающиеся в вакансиях.

### Кодирование новых признаков
[Назад к содержанию](#section_id)

Для исследования выведем несколько признаков на основе имеющихся данных для облегчения формирования групп для исследования.
У нас будет 4 категории признаков, три из которых это кодирание каждой категории из нескольких категориальных столбцов
 1. Признаки из столбцов, где в каждоый ячейке перечислено несколько значениий.

-  `city_count` - покажет число городов, в которых предполагается работа независимо от формата.
- `work_format_count` - покажет число форматов работы, предлагаемых по вакансии.
-  `job_time_count` - покажет количество форматов занятости, предлагаемых заказчиком
-  `one_city` - булевый признак. Вакансия представлена в одном городе (1) или в нескольних (0)
-  `full_salary` - булевый признак, для вакансии полностью указана зарплатная вилка.


```python
df['city_count'] = df['city'].str.split(', ').apply(lambda x: len(set(x)))
df['work_format_count'] = df['work_format'].str.split(', ').apply(lambda x: len(set(x)))
df['job_time_count'] = df['job_time_type'].str.split(', ').apply(lambda x: len(set(x)))
```

```python
df['1_one_city'] = (df['city_count'] ==  1)
df['1_full_salary'] = df['salary_max'].notnull() & df['salary_min'].notnull()
df['1_cities'] = df['city'].str.split(', ')
df['1_is_moscow'] = df['1_cities'].apply(lambda x:   1 if 'Москва' in x else   0)
df['1_is_single_moscow'] = df.apply(lambda row:  1 if row['city_count'] ==  1 and row['1_is_moscow'] ==  1 else  0, axis=1)
```

2. Кодирование столбца `job_time_type`

Если в описании вакансии встречается определенный тип формата занятости, то тогда будет присвоено значение True


```python
df['job_time'] = df['job_time_type'].str.split(', ')
df['2_full_time'] = df['job_time'].apply(lambda x: 'Полная' in x)
df['2_part_time'] = df['job_time'].apply(lambda x: 'Частичная' in x)
df['2_project_time'] = df['job_time'].apply(lambda x: 'Проектная' in x)
df['2_jun_time'] = df['job_time'].apply(lambda x: 'Стажировка' in x)
df['2_vol_time'] = df['job_time'].apply(lambda x: 'Волонтерство' in x)
```

```python
#построим и отрисуем матрицу корреляций
cm = df.drop(['id'],axis = 1).corr(method='spearman').round(2)
cm_masked = cm.mask(cm ==  1)

# #нарисуем тепловую карту с подписями для матрицы корреляций
plt.figure(figsize=(20,20))
sns.heatmap(cm_masked, 
            annot=True, 
            square=True,
            cmap='RdYlGn', 
            center = 0)
ax = plt.gca()
ax.set_title('Хитмэп матрицы корреляции');

plt.show();
```

```python
cm_df = pd.DataFrame(cm_masked)
```

```python
pairs = cm_df.unstack().reset_index()
pairs.columns = ['var_1', 'var_2', 'corr']

# Сортировка DataFrame по значениям корреляции
pairs_sorted = pairs.sort_values(by='corr', ascending=False)

# Вывод ранжированных пар
pairs_sorted

pairs_sorted['unique_pair'] = pairs_sorted.apply(lambda row: tuple(sorted([row['var_1'], row['var_2']])), axis=1)

# Удаление дубликатов по уникальному индексу
pairs_sorted = pairs_sorted.drop_duplicates(subset='unique_pair', keep='first')

# Удаление временного столбца 'unique_pair'
pairs_sorted = pairs_sorted.drop(columns=['unique_pair'])

# Вывод обновленного DataFrame
pairs_sorted
```

Далее проранжирую в таблице значения корреляций для удобства
- 0,01 < r <= 0,29 – слабая положительная связь,
- 0,30 < r <= 0,69 – умеренная положительная связь,
- 0,70 < r <= 1,00 – сильная положительная связь,
- -0,01 > r >= -0,29 – слабая отрицательная связь,
- -0,30 > r >= -0,69 – умеренная отрицательная связь,
- -0,70 > r >= -1,00 – сильная отрицательная связь. 

```python
def assign_correlation_range(correlation):
    if correlation > 0.69 and correlation <= 1:
        return 'Сильная положительная'
    elif correlation > 0.3 and correlation <= 0.69:
        return 'Умеренная положительная'
    elif correlation >= 0.01 and correlation <= 0.3:
        return 'Слабая положительная'
    elif correlation <= -0.01 and correlation >= -0.29:
        return 'Слабая отрицательная'
    elif correlation <= -0.3 and correlation >= -0.69:
        return 'Умеренная отрицательная'
    elif correlation <= -0.7 and correlation >= -1:
        return 'Сильная отрицательная'
    else:
        return 'Нулевая, полная, не определена'

pairs_sorted['cor_range'] = pairs_sorted['corr'].apply(assign_correlation_range)
```

## Проверка гипотез
[Назад к содержанию](#section_id)

Проверю несколько гипотез, связанных с вакансиями, закрытыми КА.
Для начала проверю несколько общих гипотез, а затем несколько гипотез, связанных с вакансиями, закрытыми КА.

Функция ниже позволяет провести статистические тесты на основе показателей ассимметрии и эксцесса. 

Если показатель эксцесса находится в диапазоне (-2;2), а показатель ассимметрии - (-0,5;0.5), то тогда применяется T-тест, в противном случае - тест Манна-Уитни.

Так же можно будет увидеть показатели средних и медианных значений в испытуемых группах, а так же процентную разницу значений в группах. 

```python
def hyps(df, col_1, col_2):
    a = df[df[col_1] ==  True][col_2]
    b = df[df[col_1] ==  False][col_2]
    
    skew_value = df[col_2].skew()
    kurtosis_value = df[col_2].kurtosis()
    mean_a = a.mean()
    mean_b = b.mean()
    median_a = a.median()
    median_b = b.median()
    alpha =  0.05  # критический уровень статистической значимости
    
    print('Значение ассиметрии -', skew_value)
    print('Значение эксцесса -', kurtosis_value)
    print()
    print('Среднее значение в выборке a:', mean_a)
    print('Среднее значение в выборке b:', mean_b)
    print('Разница средних в группах:', round((((a.mean()/b.mean())-1)*100),2), '%')
    print()
    print('Медианное значение в выборке a:', median_a)
    print('Медианное значение в выборке b:', median_b)
    print('Разница медиан в группах:', round((((a.median()/b.median())-1)*100),2), '%')
    print()
    # Выбор теста на основе значений kurtosis и skew
    if -2 <= kurtosis_value <=  2 or -0.5 <= skew_value <=  0.5:
        # Выполнение теста T
        results_st = st.ttest_ind(a, b, equal_var=False)
        print('Значение статистики: ', results_st.statistic)
        print('p-значение T-test:', results_st.pvalue)
        if results_st.pvalue < alpha:
            
            print("T-test: Отвергаем нулевую гипотезу")
        else:
            print("T-test: Не получилось отвергнуть нулевую гипотезу")
    else:
        # Выполнение теста U
        results_ma = st.mannwhitneyu(a, b)
        print('Значение статистики: ', results_ma.statistic)
        print('p-значение U-test: ', results_ma.pvalue)
        if results_ma.pvalue < alpha:
            print('U-test: Отвергаем нулевую гипотезу')
        else:
            print('U-test: Не получилось отвергнуть нулевую гипотезу')
```

Гипотезы будут поделены на две категории 
- анализирующие датасет в целом
- анализирующие вакансии со статусом "закрытые КА"
